## 5. Prep for Modeling

Create a DataFrame that is ready for modeling with each row representing a customer and the following numeric, non-null columns:
* Customer ID
* Whether a customer cancelled or not
* Whether a customer received a discount or not
* The number of listening sessions
* Percent of listening history consisting of Pop
* Percent of listening history consisting of Podcasts

```python
# Create a dataframe ready for modeling
cust.head()
```

```python
 model_df = cust[['Customer ID' , 'IsCancelled','Discount?']]
```

```python
listening.groupby('Customer ID')['Session ID'].nunique().rename('NumOfSessions').to_frame().reset_index()
```

```python
# Calculate the number of listening sessions for each customer
cust_sessions = listening.groupby('Customer ID')['Session ID'].nunique().rename('NumOfSessions').to_frame().reset_index()
model_df = model_df.merge(cust_sessions,how = 'left' , on = 'Customer ID')
# cust_sessions = pd.DataFrame(listening.groupby('Customer ID')['Session ID'].nunique().rename('NumOfSessions'))
#df_modeling.merge(cust_sessions,how = 'left' , on = 'Customer ID').rename(columns = {'Session ID' : 'NumOfSessions'})
```

```python
# Percent pop

listening_all = listening.merge(audio_all, how = 'left' , left_on = 'Audio ID' , right_on = 'Audio Id')
pd.get_dummies(listening_all.Genre)
```

```python
pd.concat([listening_all['Customer ID'],pd.get_dummies(listening_all.Genre)], axis = 1)
```

```python
genre = pd.concat([listening_all['Customer ID'],pd.get_dummies(listening_all.Genre)], axis = 1).groupby('Customer ID').sum()
```

```python
tot_audio = listening_all.groupby('Customer ID')['Audio Id'].count().rename('tot_audio').to_frame().reset_index()
```

```python
model_df.merge(tot_audio,how = 'left' , on ='Customer ID')
```

```python
master_audio = tot_audio.merge(genre,how = 'left' , on  = 'Customer ID')
```

```python
model_df['pop_percent'] = ( master_audio.Pop / master_audio.tot_audio ) * 100
```

```python
model_df['podcast_percent'] = ((master_audio.Comedy + master_audio['True Crime']) / master_audio.tot_audio) * 100
```
